```python
%sql select MAX(Safety_Score) AS MAX_SAFETY_SCORE from CHICAGO_PUBLIC_SCHOOLS_DATA
```

```python
%sql SELECT * FROM CHICAGO_PUBLIC_SCHOOLS_DATA WHERE Safety_Score = (select MAX(Safety_Score) AS MAX_SAFETY_SCORE from CHICAGO_PUBLIC_SCHOOLS_DATA);
```

```python
%sql select Name_of_School, Average_Student_Attendance from CHICAGO_PUBLIC_SCHOOLS_DATA \
    order by Average_Student_Attendance desc nulls last limit 10 
```

```python
%sql select Name_of_School, Average_Student_Attendance from CHICAGO_PUBLIC_SCHOOLS_DATA \
    order by Average_Student_Attendance nulls last limit 5; 
```

```python
%sql SELECT Name_of_School, REPLACE(Average_Student_Attendance, '%', '') \
     from CHICAGO_PUBLIC_SCHOOLS_DATA \
     order by Average_Student_Attendance \
     LIMIT 5
```

```python
%sql SELECT * from CHICAGO_PUBLIC_SCHOOLS_DATA where Average_Student_Attendance < 70
```

```python
%sql SELECT Community_Area_Name, SUM(COLLEGE_ENROLLMENT) FROM CHICAGO_PUBLIC_SCHOOLS_DATA \
    GROUP BY Community_Area_Name 
```

```python
%sql select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT \
   from CHICAGO_PUBLIC_SCHOOLS_DATA \
   group by Community_Area_Name order by TOTAL_ENROLLMENT limit 5;
```

```python
%sql select NAME_OF_SCHOOL, Safety_Score FROM CHICAGO_PUBLIC_SCHOOLS_DATA where Safety_Score != 'None' order by Safety_Score limit 5;
```
